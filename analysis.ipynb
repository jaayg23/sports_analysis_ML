```python
import pandas as pd
```

```python
club_data = pd.read_csv('club_data.csv')
club_data.head()
```

```python
club_data.info()
```

```python
club_data.isnull().sum()
club_data['goles_local'] = club_data['goles_local'].fillna(club_data['goles_local'].mean())
print(club_data.isnull().sum())
```

```python
#One hot encoding
prepared_data = pd.get_dummies(club_data, columns=['equipo_local', 'equipo_visitante'])
prepared_data.head()
```

```python
#Eliminating duplicate
print('Rows before removing duplicates:', len(prepared_data))
prepared_data.drop_duplicates(inplace=True)
print('Rows after removing duplicates:', len(prepared_data))
```

```python
prepared_data['fecha_partido'] = pd.to_datetime(prepared_data['fecha_partido'], errors='coerce')
```
