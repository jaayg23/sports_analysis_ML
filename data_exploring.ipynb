```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_matches = pd.read_csv('club_data.csv')
df_matches.head()
```

```python
df_matches.describe()
```

```python
goals_local = df_matches[df_matches['equipo_local'] == 'Cebollitas FC']['goles_local'].mean()
goals_visitante = df_matches[df_matches['equipo_visitante'] == 'Cebollitas FC']['goles_visitante'].mean()
print(f"Average goals scored by Cebollitas FC at home: {goals_local}")
print(f"Average goals scored by Cebollitas FC away: {goals_visitante}")
```

```python
fig, ax = plt.subplots(1,2, figsize=(12, 4))

sns.histplot(df_matches['goles_local'], kde=True, ax=ax[0], bins=10)
ax[0].set_title('Distribution of Goals Scored by Home Team')

sns.histplot(df_matches['goles_visitante'], kde=True, ax=ax[1], bins=10, color='orange')
ax[1].set_title('Distribution of Goals Scored by Away Team')

plt.show()
```

```python
sns.boxplot(x='goles_local', data=df_matches, color='lightblue')
plt.title('Boxplot of Goals Scored by Home Team')
plt.xlabel('Goals Scored by Home Team')
plt.show()
```

```python
sns.scatterplot(x='posesion_local (%)', y='goles_local', data=df_matches, alpha=0.5)
plt.title('Scatter Plot of Home Team Possession vs Goals Scored')
plt.xlabel('Home Team Possession (%)')
plt.ylabel('Goals Scored by Home Team')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df_matches[['goles_local', 'goles_visitante', 'posesion_local (%)',
                        'posesion_visitante (%)', 'tiros_arco_local',
                        'tiros_arco_visitante']].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Match Features')
plt.show()
```
