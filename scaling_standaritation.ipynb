```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
df_matches = pd.read_csv('club_data.csv')
df_matches.head()
```

```python
scaler_norm = MinMaxScaler()

df_matches['tiros_arco_local_norm'] = scaler_norm.fit_transform(df_matches[['tiros_arco_local']])
df_matches['tiros_arco_visitante_norm'] = scaler_norm.fit_transform(df_matches[['tiros_arco_visitante']])

print(df_matches[['tiros_arco_local','tiros_arco_local_norm']].head(6))
df_matches[['tiros_arco_visitante','tiros_arco_visitante_norm']].head(6)
```

```python
scaler_std = StandardScaler()
df_matches['posesion_local_std'] = scaler_std.fit_transform(df_matches[['posesion_local (%)']])
df_matches['posesion_visitante_std'] = scaler_std.fit_transform(df_matches[['posesion_visitante (%)']])

print(df_matches[['posesion_local (%)', 'posesion_local_std']].head(6))
df_matches[['posesion_visitante (%)', 'posesion_visitante_std']].head(6)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(1, 2, figsize=(12, 4))

sns.histplot(df_matches['tiros_arco_local_norm'], bins=10, kde=True, ax=ax[0], color='blue', label='Local')
ax[0].set_title('Tiros al arco Local (Normalizado)')

sns.histplot(df_matches['posesion_local_std'], bins=10, kde=True, ax=ax[1], color='blue', label='Local')
ax[1].set_title('Posesión Local (Estandarizado)')

plt.show()
```
